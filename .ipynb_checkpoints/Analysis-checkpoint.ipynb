```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('final.csv',  parse_dates=['date'], dtype={'name':'category'})
data.head()
```

```python
data.info()
```

```python
data.groupby('name').describe()
```

```python
data['time'] = data['date'].dt.round('15min').dt.time
data.tail()
```

```python
# plotting; converting the dataset to wide format
data_wide = data.pivot("name", "time", "val")
data_wide.head()

sns.set()
cmap = sns.cm.rocket_r
f, ax = plt.subplots(figsize=(15, 13))
sns.heatmap(data_wide, annot=False, linewidths=.3, ax=ax, cmap = cmap, annot_kws={"size": 10})
```

```python
def dateplot(x, y, **kwargs):
    ax = plt.gca()
    data = kwargs.pop("data")
    data.plot(x=x, y=y, ax=ax, grid=True, **kwargs)
    
g = sns.FacetGrid(data, col="name", col_wrap=4, height=4, sharex=False)
g = g.map_dataframe(dateplot, "time", "val")
```
